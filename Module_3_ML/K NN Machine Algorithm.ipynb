The dataset choosen for this experiment is  Wisconsin Breast Cancer data to detect malignant cells.The dataset has 569 samples and 32 columns.

## **Exercises**

### **Exercise 1**

**Importing Necessary Libararies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

### **Exercise 2**

**Loading the dataset**

```python
link = "/content/Breast_cancer_data.csv"
data_set = pd.read_csv(link)
data_set.head()
```

### **Exercise 3**

**Split the dataset into training and testing data** 

```python
features = data_set.drop(['Unnamed: 32', 'id', 'diagnosis'], axis=1)
target = data_set['diagnosis']

X_train , X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
for each in [X_train, X_test, y_train, y_test]:
    print(each.shape)
```

```python
X_train
```

### **Exercise 4**

**Using all the attributes/features, predict the patients in the testdata as malignant or benign. Use K = 5 and Euclidean distance. Find the accuracy (as percentage) on the test data.**

```python
knn_model = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_model.fit(X_train, y_train)

test_preds = knn_model.predict(X_test)

print(f"Testing Accuracy: {accuracy_score(y_test, test_preds) * 100:.2f}%")
```

### **Exercise 5** 
**Vary K from 3 to 11 and find the best K.**

```python
k_range = range(3, 12)
test_acc = {}
for k in k_range:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train, y_train)
    test_preds = knn_model.predict(X_test)

    test_acc[k] = accuracy_score(y_test, test_preds)
test_acc
```

```python
best_k, accuracy = sorted(test_acc.items(),
key=lambda x: x[1], reverse=True)[0]

print(f"Best K value: {best_k}\nTest Accuracy: {accuracy * 100:.2f}%")
```
